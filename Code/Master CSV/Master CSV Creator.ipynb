```python
sheets = ['Database User', 'Database ZEV', 'Database BYD', 'Database RYD', 'Database Fuel']
#path = os.getcwd()+'/'+'Forms'
path = 'C:\\Users\\tsingleton\\Desktop\\CALSTART\\2. Sus Fleets\\4. Version Control Github\\SF-Github-Repo\\Report\\Fleet Enrollment Forms'
```

```python
for file in os.listdir(path):
    print(file)
    #for the current file create a separate dataframe for each database sheet and store it in a list
    for sheet in sheets:
        if sheet == 'Database User':
            df = pd.read_excel(path+'/'+file, sheet_name = sheet)
            df.rename({0: 'fleet', 1: 'size', 2: 'baseline_year', 3: 'reporting_year', 4: 'units', 5: 'description', 6: 'other'},axis = 1)
            df.to_csv('Master Sheets/Master User.csv', index = None, header = None, sep = ' ', mode = 'a')
            
        if sheet == 'Database ZEV':
            df = pd.read_excel(path+'/'+file, sheet_name = sheet)
            df['fleet'] = file.split(' Sustainable ')[0]
            df.to_csv('Master Sheets/Master ZEV.csv', index = None, header = None, sep = ' ', mode = 'a')
            
        if sheet == 'Database BYD':
            df = pd.read_excel(path+'/'+file, sheet_name = sheet)
            df['fleet'] = file.split(' Sustainable ')[0]
            df.to_csv('Master Sheets/Master BYD.csv', index = None, header = None, sep = ' ', mode = 'a')
            
        if sheet == 'Database RYD':
            df = pd.read_excel(path+'/'+file, sheet_name = sheet)
            df['fleet'] = file.split(' Sustainable ')[0]
            df.to_csv('Master Sheets/Master RYD.csv', index = None, header = None, sep = ' ', mode = 'a')
            
        if sheet == 'Database Fuel':
            df = pd.read_excel(path+'/'+file, sheet_name = sheet)
            df['fleet'] = file.split(' Sustainable ')[0]
            df.to_csv('Master Sheets/Master Fuel.csv', index = None, header = None, sep = ' ', mode = 'a')
```
